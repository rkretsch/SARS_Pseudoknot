```python
# function that outputs a list of sliding windows in a given sequence
# arguments: given sequence, start position, end position, step (number of sliding nucleotides), window length
# will print a warning if 
def get_sliding_windows(sequence_name, start, end, step, window):
    windows = []
    for i in range(start, end, step):
        new_windows = sequence_name[i:i+window]
        if len(new_windows) == window:
            windows.append(new_windows)
    return windows
```

```python
def get_sliding_windows_output(sequence_name, start, end, step, window):
    windows = []
    for i in range(start, end, step):
        new_windows = sequence_name[i:i+window]
        if len(new_windows) == window:
            windows.append(new_windows)
    windows_string = str(windows)
    file = open("sliding_windows_output.txt", "w")
    file.writelines(windows_string)
    file.close()
```
